```python
import pandas as pd
```

## 1. Review and Clean Dataset

### 1.1 Dataset Loading and Initial Review

```python
# Read the loaded dataset
df=pd.read_csv("US_Crime_Data.csv")
```

### 1.2 Data cleaning operations

```python
# Convert the 'Date' column to datetime type.
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# Manage missing data:
# Fill in the missing data in the 'City' and 'State' columns with 'Unknown'.
df["City"].fillna("Unknown", inplace=True)
df["State"].fillna("Unknown", inplace=True)
```

```python
# Fill in the missing data in the 'Keyword' and 'Summary' columns with an empty string ('').
df["Keyword"].fillna('', inplace=True)
df["Summary"].fillna('', inplace=True)
```

```python
# Remove the repetitive ones.
df.drop_duplicates(inplace=True)
```

```python
# Standardize the category columns:
# Convert the data in the 'City' and 'State' columns to uppercase.
df["City"] = df["City"].str.upper()
df["State"] = df["State"].str.upper()
```

### 1.3 Data Set After Changes

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
from sklearn.impute import SimpleImputer
```

```python
# Missing data management (filling with the most frequently repeated value)
imputer = SimpleImputer(strategy="most_frequent")
df["City"] = imputer.fit_transform(df["City"].values.reshape(-1, 1)).ravel()
df["State"] = imputer.fit_transform(df["State"].values.reshape(-1, 1)).ravel()
```

```python
# Make a new features from the data set: Year, Month and Day
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month_name()
df["Day"] = df["Date"].dt.dayofweek  # Monday to Sunday: 0 to 6
```

```python
df.head()
```

## 2. Data Processing

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
# First, use the 'State' column as the target variable
# Separate the target variable and properties
X = df.drop(["State"], axis=1)
y = df["State"]
```

```python
# Since the 'State' column is categorical, convert it to a numeric format
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
# Divide the data set into training and testing set (e.g. 70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X["Title"], y_encoded, test_size=0.3, random_state=42)
```

```python
# Fill the NaN values in the 'Title' column with empty string ('')
df["Title"].fillna('', inplace=True)
```

## 3. Model Train

```python
# Split the data set into training and test set (this time NaN values for the 'Title' column will not cause a problem)
X_train, X_test, y_train, y_test = train_test_split(df["Title"], y_encoded, test_size=0.3, random_state=42)
```

```python
# Do a simple vectorization process
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words="english", max_features=1000)
X_train_transformed = vectorizer.fit_transform(X_train)
X_test_transformed = vectorizer.transform(X_test)
```

```python
# Build and train the Logistic Regression model
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train_transformed, y_train)
```

```python
# Build and train the Random Forest model
random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_model.fit(X_train_transformed, y_train)
```

```python
# Evaluate the performance of the models
logistic_preds = logistic_model.predict(X_test_transformed)
rf_preds = random_forest_model.predict(X_test_transformed)
```

```python
logistic_accuracy = accuracy_score(y_test, logistic_preds)
rf_accuracy = accuracy_score(y_test, rf_preds)
```

```python
logistic_report = classification_report(y_test, logistic_preds)
rf_report = classification_report(y_test, rf_preds)
```

```python
# Do n-gram and sentiment analysis for text processing
tfidf_ngram_vectorizer = TfidfVectorizer(stop_words="english", ngram_range=(1,2), max_features=2000)
text_ngram_features = tfidf_ngram_vectorizer.fit_transform(text_features).toarray()
```

```python
#pip install textblob
```

```python
# TextBlob for sentiment analysis
from textblob import TextBlob
df["Sentiment"] = df["Summary"].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
# Process categorical features with one-hot encoding
encoder = OneHotEncoder(sparse=False)
encoded_features = encoder.fit_transform(df[["City", "Organization"]])
```

```python
# Create interaction properties (a simple example such as the combination of City and Organization)
df["City_Org"] = df["City"] + "_" + df["Organization"]
interaction_encoder = OneHotEncoder(sparse=False)
interaction_features = interaction_encoder.fit_transform(df[["City_Org"]])
```

```python
# All the features together
all_features = np.hstack((text_ngram_features,
                          df[["Year", "Month", "Day", "Weekday", "Sentiment"]].values,
                          encoded_features,
                          interaction_features))
```

```python
# Separate the data set into training and test set
X_train, X_test, y_train, y_test = train_test_split(all_features, y_encoded, test_size=0.3, random_state=42)
```

```python
# Build and train the Random Forest model
random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_model.fit(X_train, y_train)
```

```python
# Evaluate the performance of the model
rf_preds = random_forest_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_preds)
rf_report = classification_report(y_test, rf_preds)
```

## 4. Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Defining the model
rf = RandomForestClassifier(random_state=42)
```

```python
# Define the grid of parameters to test
param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [None, 20],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2]
}
```

```python
random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_grid, n_iter=10, cv=2, verbose=2, random_state=42, n_jobs=1)
```

```python
# Run grid search
random_search.fit(X_train, y_train)
```

```python
# Print the best parameters and score
print("Best Parameters:", random_search.best_params_)
print("Best Score:", random_search.best_score_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define the model with the best parameters determined before
model = RandomForestClassifier(n_estimators=100, min_samples_split=5, min_samples_leaf=1, max_depth=None, random_state=42)
```

```python
# Train the model on all training data
model.fit(X_train, y_train)
```

```python
# Calculate crime rates by months and states
df["Month"] = df["Date"].dt.month_name()
monthly_state_crimes = df.groupby(["Month", "State"]).size().unstack(fill_value=0)
```

```python
# Calculating ratios with normalized data
monthly_state_rates = monthly_state_crimes.div(monthly_state_crimes.sum(axis=0), axis=1) * 100
```

```python
# Sort by month
months_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
monthly_state_rates = monthly_state_rates.reindex(months_order)
```

```python
# Visualization
plt.figure(figsize=(12, 8))
sns.heatmap(monthly_state_rates, cmap="coolwarm", annot=True, fmt=".1f", linewidths=.5)
plt.title("Percentage of Crimes by State Across Months")
plt.ylabel("Month")
plt.xlabel("State")
plt.show()
```
